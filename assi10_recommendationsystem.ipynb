```python
book=book.iloc[:,1:]
book.head()
```

```python
len(book['Book.Title'].unique())
```

```python
book1=book.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
book1.head()
```

```python
book1.index=book['User.ID'].unique()
book1
```

```python
book1.fillna(0,inplace=True)
book1
```

```python
#calculating cosine similarity
```

```python
user_sim=1-pairwise_distances(book1.values,metric='cosine')
user_sim
```

```python
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
user_sim2.index=book['User.ID'].unique()
user_sim2.columns=book['User.ID'].unique()
user_sim2
```

```python
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
user_sim2.idxmax(axis=1)
```
